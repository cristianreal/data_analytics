```python
import pandas as pd
df=pd.read_parquet('parquet/eventos_transformados.parquet.gzip')  
df
```

```python
df = df[df["event"]=="transaction"]
df
```

```python
transaction_category_counts = df.groupby('transactionid')['categoria'].nunique()

# Step 2: Filter for transactions with more than 2 categories
transactions_with_more_than_2 = transaction_category_counts[transaction_category_counts > 2]

df_reglas_asociacion = df[["transactionid","categoria"]]

df_reglas_asociacion.sort_values(by="transactionid",ascending=True).head()
```

```python
basket_sizes = df.groupby('transactionid').size()
valid_baskets = basket_sizes[basket_sizes > 1].index
df_filtered = df[df['transactionid'].isin(valid_baskets)]
df_filtered.sort_values(by="transactionid",ascending=True)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Step 2: Create a one-hot encoded DataFrame
one_hot = pd.get_dummies(df_filtered['itemid'])
one_hot['transactionid'] = df_filtered['transactionid']
basket = one_hot.groupby('transactionid').max()

# Step 3: Generate frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.001, use_colnames=True)

# Step 4: Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Display the rules

# Optional: Check the number of transactions before and after filtering
print(f"Original number of transactions: {df['transactionid'].nunique()}")
print(f"Number of transactions after removing single-item baskets: {df_filtered['transactionid'].nunique()}")

rules.sort_values(by="support",ascending=False)

```

```python


from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df2.astype('bool'), min_support=0.001, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift")

rules.head()
```

```python
rules.sort_values(["support", "confidence","lift"],axis = 0, ascending = False)
```

```python
rules.to_csv('csv/carritos_transacciones_reglas_asociacion_categoria.csv')  
```

```python
from sqlalchemy import create_engine
import numpy as np

engine = create_engine('postgresql://data_analytics:data_analytics@127.0.0.1:5432/data_analytics')
rules['new_antecedents'] = rules['antecedents'].apply(lambda x: np.array(list(x)))
rules['new_antecedents'] = rules['new_antecedents'].astype(str)
rules['new_consequents'] = rules['consequents'].apply(lambda x: np.array(list(x)))
rules['new_consequents'] = rules['new_consequents'].astype(str)

rules[["new_antecedents","new_consequents","antecedent support","consequent support","support","confidence","lift"]].to_sql('reglas_asociacion_categorias', engine, if_exists='replace', index=False)
rules
```
